```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
import scipy.stats as stats
```

```python
np.random.seed(42)
# plot some normally distributed data
data = np.random.normal(100, 15, 3000)
df = pd.DataFrame({'normal_data': data})
sns.set_theme(style="ticks", palette="pastel")

plt.figure(figsize=(5, 3))
sns.histplot(df['normal_data'], kde=True, bins=30)
plt.savefig('plots/normal_distribution_plot.png', bbox_inches='tight', dpi=300)
plt.show()


# plot some skewed and bimodal distributed data

skewed_data = np.random.exponential(2, 1250)
skewed_df = pd.DataFrame({'skewed_data': skewed_data})

bimodal_data = np.concatenate([np.random.normal(75, 10, 500), np.random.normal(125, 25, 750)])
bimodal_df = pd.DataFrame({'bimodal_data': bimodal_data})


plt.figure(figsize=(5, 3))
sns.histplot(skewed_df['skewed_data'], kde=True, bins=30)
plt.tight_layout()
plt.savefig('plots/skewed_distribution_plot.png', bbox_inches='tight', dpi=300)
plt.show()

plt.figure(figsize=(5, 3))
sns.histplot(bimodal_df['bimodal_data'], kde=True, bins=30)
plt.tight_layout()
plt.savefig('plots/bimodal_distribution_plot.png', bbox_inches='tight', dpi=300)
plt.show()
```
